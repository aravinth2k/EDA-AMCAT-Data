```python
from fuzzywuzzy import process

def spell_check(target='', choice=[],threashold=65):
    word, theres = process.extractOne(target,choice)
    if theres >= threashold:
        return word
    else:
        return 'Other'
```

```python
df['JobCity2'] = df['JobCity'].apply(lambda x: spell_check(str(x),choices))
```

```python
def board(data):
    if ('cbs' in data.lower()) or ('central' in data.lower()):
        return 'CBSE'
    elif 'mat' in data.lower():
        return 'Matriculation'
    elif 'icse' in data.lower():
        return 'ICSE'
    else:
        return 'State Board'
```

```python
df['10board'] = df['10board'].apply(lambda x: board(str(x)))
```

```python
df['12board'] = df['12board'].apply(lambda x: board(str(x)))
```

```python
def age(dob):
    today = pd.to_datetime(maxdate)
    return today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
```

```python
df['age'] = df['DOB'].apply(age)
```
