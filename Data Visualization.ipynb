```python
filter = df.groupby(['Specialization'])['Salary'].count().sort_values(ascending=False)[:20].index

def check(i):
    if i in filter:
        return True
    else:
        return False

temp = df[df['Specialization'].apply(check)]
```

```python
pd.crosstab(index=temp['Specialization'],columns=temp['Gender'],values=temp['Specialization'],aggfunc='count')[:20].plot(kind='barh', stacked=True)
```

```python
cat_var = [('Gender','Specialization')]
```

```python
#Chi square Test
import scipy.stats as ss
result = []
for i in cat_var:
    if i[0] != i[1]:
        result.append((i[0],i[1],list(ss.chi2_contingency(pd.crosstab(df[i[0]], df[i[1]])))[1]))
print('Chi-Square Test Value')
result
```
